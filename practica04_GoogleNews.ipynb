```python
import stanza
import pandas as pd
import json

# Download the Stanza model (if you haven't already) - Run this ONCE
stanza.download('en')

# Initialize the Stanza pipeline
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,lemma,depparse,ner')  # Include necessary processors

def analyze_news(df, text_column):
    """
    Performs linguistic analysis of news articles using Stanza.
    Includes:
    - Sentence segmentation
    - Tokenization
    - POS tagging
    - Lemmatization
    - Dependency parsing
    - Named Entity Recognition (NER)
    """
    analyzed_news = []

    for text in df[text_column]:
        doc = nlp(text)
        analysis = {
            "sentences": [sentence.text for sentence in doc.sentences],
            "tokens": [[word.text for word in sentence.words] for sentence in doc.sentences],
            "pos_tags": [[(word.text, word.upos, word.xpos) for word in sentence.words] for sentence in doc.sentences],
            "lemmas": [[word.lemma for word in sentence.words] for sentence in doc.sentences],
            "dependencies": [[(word.text, word.head, word.deprel) for word in sentence.words] for sentence in doc.sentences],
            "entities": [[(ent.text, ent.type) for ent in sentence.ents] for sentence in doc.sentences]
        }
        analyzed_news.append(analysis)

    return analyzed_news

# --- 1. Load your Google News data (ADAPT THIS PART) ---
# (Replace this with your actual Google News data loading)

try:
    google_news_texts = googlenews.get_texts()  # Replace with your actual function
    df_news = pd.DataFrame({'text': google_news_texts})
except NameError:  # googlenews is not defined
    print("Please, define the googlenews.get_texts() function or replace it with your actual way of retrieving the text data")
    exit()
except AttributeError:  # googlenews does not have the method get_texts
    print("Please, verify that the googlenews object has a 'get_texts()' method or replace it with your actual way of retrieving the text data")
    exit()

# --- 2. Analyze the news data ---
analyzed_news = analyze_news(df_news, 'text')

# --- 3. Print the analyzed news data to console ---
for analysis in analyzed_news:
    print(analysis)  # Or use json.dumps() for better formatting

# --- 4. Save the analyzed news data to a JSON file ---
with open('google_news_analysis.json', 'w', encoding='utf-8') as f:
    json.dump(analyzed_news, f, ensure_ascii=False, indent=4)
```

```python
import stanza
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
import json
import random

# Download the Stanza model (if you haven't already)
stanza.download('en')

# Initialize the Stanza pipeline
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,lemma')

def preprocess_text(text):
    """Preprocesses a single text string using Stanza."""
    doc = nlp(text)
    lemmas = [word.lemma for sent in doc.sentences for word in sent.words]
    return ' '.join(lemmas)

def train_classifier(texts, labels):
    """Trains and evaluates a text classification model."""
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform(texts)
    X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

    classifier = LogisticRegression()
    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    print("Classifier Evaluation:")
    print(f"Accuracy: {accuracy}")
    print(report)

    return vectorizer, classifier

def classify_new_text(text, vectorizer, classifier):
    """Classifies new text using the trained model."""
    preprocessed_text = preprocess_text(text)
    vectorized_text = vectorizer.transform([preprocessed_text])
    prediction = classifier.predict(vectorized_text)[0]
    return prediction

# --- 1. Load your Google News data (ADAPT THIS PART) ---
try:
    google_news_texts = googlenews.get_texts()  # Replace with your actual function
    df_news = pd.DataFrame({'text': google_news_texts})
except NameError:  # googlenews is not defined
    print("Please, define the googlenews.get_texts() function or replace it with your actual way of retrieving the text data")
    exit()
except AttributeError:  # googlenews does not have the method get_texts
    print("Please, verify that the googlenews object has a 'get_texts()' method or replace it with your actual way of retrieving the text data")
    exit()

# --- 2. Add Labels to your data ---
labels = [random.randint(0, 1) for _ in range(len(df_news))]
df_news['label'] = labels

# --- 3. Preprocess the text data ---
df_news['text'] = df_news['text'].apply(preprocess_text)

# --- 4. Train the classifier ---
vectorizer, classifier = train_classifier(df_news['text'], df_news['label'])

# --- 5. Example Usage ---
new_text = googlenews.get_texts()[0] # Replace with your actual text
prediction = classify_new_text(new_text, vectorizer, classifier)
print(f'Prediction for new text: {prediction}')

# --- 6. (Optional) Store the data in a JSON file ---
data_to_save = df_news.to_dict(orient='records')
with open('google_news_classification.json', 'w', encoding='utf-8') as f:
    json.dump(data_to_save, f, ensure_ascii=False, indent=4)

print("\nData saved to google_news_classification.json")
```

```python
import stanza
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Download Stanza model for English (if you haven't already)
stanza.download('en')

# Initialize Stanza pipeline with only the necessary processors
nlp = stanza.Pipeline('en', processors='tokenize,lemma')

def preprocess_text(text):
    """Preprocesses text using Stanza for Information Retrieval."""
    doc = nlp(text)
    lemmas = [word.lemma for sent in doc.sentences for word in sent.words if not word.text.isspace()]  # Skip whitespace lemmas
    return " ".join(lemmas)

def create_inverted_index(texts):
    """Creates an inverted index for efficient information retrieval."""
    inverted_index = {}
    for i, text in enumerate(texts):
        for word in text.split():  # Assumes texts are preprocessed
            if word: # skip empty string
                if word not in inverted_index:
                    inverted_index[word] = [i]
                elif i not in inverted_index[word]:  # Add document index if not already present
                    inverted_index[word].append(i)
    return inverted_index

def search_news(query, inverted_index, texts, top_n=5):
    """Searches for relevant news articles using the inverted index and cosine similarity."""
    query = preprocess_text(query)  # Preprocess the query
    query_terms = query.split()

    # Candidate document retrieval using the inverted index
    candidate_docs = set()
    for term in query_terms:
        if term in inverted_index:
            candidate_docs.update(inverted_index[term])

    if not candidate_docs:
      return [] # or raise an exception since no results were found.

    # Calculate TF-IDF vectors and cosine similarity (if candidates are found)
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([texts[i] for i in candidate_docs] + [query])  # Include the query in matrix

    query_vector = tfidf_matrix[-1]  # Last row is the query vector
    document_vectors = tfidf_matrix[:-1]  # All other rows are document vectors

    similarities = cosine_similarity(query_vector, document_vectors)[0]

    # Rank documents by similarity
    ranked_docs = sorted(enumerate(candidate_docs), key=lambda x: similarities[x[0]], reverse=True)
    top_results = [(texts[doc_idx], similarities[i]) for i, doc_idx in ranked_docs[:top_n]]
    return top_results

# --- 1. Load your Google News data ---
try:
    google_news_texts = googlenews.get_texts()
    df_news = pd.DataFrame({'text': google_news_texts})
except NameError:
    print("Please define googlenews.get_texts() or replace it.")
    exit()
except AttributeError:
    print("Verify googlenews has a 'get_texts()' method.")
    exit()

# --- 2. Preprocess the news articles ---
df_news['processed_text'] = df_news['text'].apply(preprocess_text)

# --- 3. Create the inverted index ---
inverted_index = create_inverted_index(df_news['processed_text'].tolist())

# --- 4. Information Retrieval Examples ---
queries = [
    "bribery",
    "overpricing",
    "robbery"
]

for query in queries:
    results = search_news(query, inverted_index, df_news['processed_text'].tolist())
    print(f"\nQuery: {query}")
    if results:
        for i, (text, similarity) in enumerate(results):
            print(f"  Result {i+1} (Similarity: {similarity:.4f}):")
            print(text[:200] + "...")  # Print a snippet
    else:
        print("No results found.")

# --- 5. (Optional) Save data to JSON ---
data_to_save = df_news.to_dict(orient='records')
with open('google_news_retrieval.json', 'w', encoding='utf-8') as f:
    json.dump(data_to_save, f, ensure_ascii=False, indent=4)

print("\nData (including processed text) saved to google_news_retrieval.json")
```

```python
import stanza
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json
from collections import Counter
from transformers import pipeline

# Download Stanza model for English (if you haven't already)
stanza.download('en')

# Initialize Stanza pipeline
nlp_stanza = stanza.Pipeline('en', processors='tokenize,lemma,pos,ner,depparse')

# Initialize Hugging Face token classification pipeline
classifier = pipeline("token-classification", model="vblagoje/bert-english-uncased-finetuned-pos")

def preprocess_text_stanza(text):
    """Preprocesses text using Stanza."""
    doc = nlp_stanza(text)
    lemmas = [word.lemma for sent in doc.sentences for word in sent.words if not word.text.isspace()]
    return " ".join(lemmas)

def create_inverted_index(texts):
    """Creates an inverted index."""
    inverted_index = {}
    for i, text in enumerate(texts):
        for word in text.split():
            if word:  # skip empty string
                if word not in inverted_index:
                    inverted_index[word] = [i]
                elif i not in inverted_index[word]:
                    inverted_index[word].append(i)
    return inverted_index

def search_news(query, inverted_index, texts, top_n=5):
    """Searches for relevant news articles."""
    query = preprocess_text_stanza(query)
    query_terms = query.split()

    candidate_docs = set()
    for term in query_terms:
        if term in inverted_index:
            candidate_docs.update(inverted_index[term])

    if not candidate_docs:
        return []  # Return empty list if no candidates

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([texts[i] for i in candidate_docs] + [query])

    query_vector = tfidf_matrix[-1]
    document_vectors = tfidf_matrix[:-1]

    similarities = cosine_similarity(query_vector, document_vectors)[0]  # Extract the array

    ranked_docs_indices = sorted(range(len(similarities)), key=lambda i: similarities[i], reverse=True)

    candidate_docs_list = list(candidate_docs)  # For indexing
    top_results = [(texts[candidate_docs_list[i]], similarities[i]) for i in ranked_docs_indices[:top_n]]

    return top_results

def analyze_sentiment_stanza(text):
    """Analyzes sentiment with Stanza."""
    doc = nlp_stanza(text)
    sentiments = [sent.sentiment for sent in doc.sentences]
    valid_sentiments = [s for s in sentiments if s is not None]  # Filter out None values
    if valid_sentiments:
        overall_sentiment = sum(valid_sentiments) / len(valid_sentiments)
    else:
        overall_sentiment = 1  # Neutral if no valid sentiments
    return overall_sentiment

def extract_named_entities_stanza(text):
    """Extracts named entities with Stanza."""
    doc = nlp_stanza(text)
    entities = []
    for ent in doc.ents:
        entities.append({"text": ent.text, "type": ent.type})
    return entities

def analyze_pos_tags_stanza(text):
    """Analyzes part-of-speech tags with Stanza."""
    doc = nlp_stanza(text)
    pos_counts = Counter()
    lemma_counts = Counter()
    for sent in doc.sentences:
        for word in sent.words:
            pos_counts[word.upos] += 1
            lemma_counts[word.lemma] += 1
    return pos_counts, lemma_counts

def classify_tokens_transformers(text):
    """Classifies tokens with Hugging Face Transformers."""
    try:
        result = classifier(text)
        return result  # Return the list of token classifications
    except Exception as e:
        print(f"Error classifying tokens: {e}")
        return None

# --- 1. Load your Google News data (REPLACE THIS) ---
try:
    google_news_texts = googlenews.get_texts()  # Placeholder
    df_news = pd.DataFrame({'text': google_news_texts})
except NameError:
    print("Error: googlenews.get_texts() is not defined.")
    exit()
except AttributeError:
    print("Error: googlenews object does not have 'get_texts()' method.")
    exit()
except Exception as e:
    print(f"An error occurred while loading data: {e}")
    exit()

# --- 2. Preprocess & Create Inverted Index ---
df_news['processed_text'] = df_news['text'].apply(preprocess_text_stanza)
inverted_index = create_inverted_index(df_news['processed_text'].tolist())

# --- 4. Information Retrieval and Token Classification ---
query = "corruption"  # Example query
results = search_news(query, inverted_index, df_news['processed_text'].tolist())

print(f"\nQuery: {query}")

all_results_data = []  # For JSON

if results:
    for i, (text, similarity) in enumerate(results):
        print(f"  Result {i+1} (Similarity: {similarity:.4f}):")
        print(text[:200] + "...")  # Snippet

        # Stanza Analysis
        sentiment_stanza = analyze_sentiment_stanza(text)
        entities_stanza = extract_named_entities_stanza(text)
        pos_counts_stanza, lemma_counts_stanza = analyze_pos_tags_stanza(text)

        # Transformers Token Classification
        token_classification_results = classify_tokens_transformers(text)
        if token_classification_results:
            print("   Token Classification (Transformers):", token_classification_results)
        else:
            print("   Token Classification (Transformers): Classification failed for this text.")

        all_results_data.append({  # Gather data for JSON
            "text": text,
            "similarity": similarity,
            "sentiment_stanza": sentiment_stanza,
            "entities_stanza": entities_stanza,
            "pos_counts_stanza": pos_counts_stanza,
            "lemma_counts_stanza": lemma_counts_stanza,
            "token_classification_results": token_classification_results
        })
else:
    print("No results found.")

# --- 5. Save data to JSON ---
with open('google_news_analysis_bert.json', 'w', encoding='utf-8') as f:
    json.dump(all_results_data, f, ensure_ascii=False, indent=4, default=str)

print("\nData (including analysis) saved to google_news_analysis_bert.json")
```
