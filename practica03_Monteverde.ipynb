```
import stanza
import json
from collections import Counter

# Descargar el modelo en español para Stanza
stanza.download('en')

# Inicializar el pipeline de Stanza con más procesadores
nlp = stanza.Pipeline('es', processors='tokenize,mwt,pos,lemma,ner,depparse,sentiment')

# Textos de las biografías (Asegúrate de que doc1, doc2 y doc3 estén definidos)
texto1 = doc1  # Democrito
texto2 = doc2  # Platon
texto3 = doc3  # Aristoteles

textos = [texto1, texto2, texto3]

# Procesar cada biografía
for i, texto in enumerate(textos):
    print(f"Procesando biografía {i+1}...")

    # Anotar el texto con Stanza
    doc = nlp(texto)

    # Extraer información semántica
    relaciones = []
    for sentence in doc.sentences:
        # Relaciones de dependencia
        for word in sentence.words:
            relacion = word.deprel
            head = sentence.words[word.head - 1].text if word.head > 0 else "ROOT"
            dependent = word.text
            relaciones.append({
                "type": "dep",
                "head": head,
                "relation": relacion,
                "dependent": dependent
            })

        # Entidades nombradas
        for ent in sentence.ents:
            relaciones.append({
                "type": "ner",
                "text": ent.text,
                "type": ent.type
            })

        # Sentimiento de la oración
        relaciones.append({
            "type": "sentiment",
            "sentence": sentence.text,
            "sentiment": sentence.sentiment
        })

    # Contar la frecuencia de las relaciones de dependencia
    conteo_relaciones_dep = Counter(rel["relation"] for rel in relaciones if rel["type"] == "dep")

    # Imprimir resultados
    print(f"Relaciones de Dependencia con Frecuencia para biografía {i+1}:")
    for relacion, conteo in conteo_relaciones_dep.most_common():
        print(f"{relacion}: {conteo}")

    print(f"\nInformación Semántica Detallada para biografía {i+1}:")
    for rel in relaciones:
        print(json.dumps(rel, indent=4, ensure_ascii=False))

    # Guardar las relaciones en un archivo JSON
    nombre_archivo = f"relaciones_biografia_{i+1}.json"
    with open(nombre_archivo, 'w', encoding='utf-8') as f:
        json.dump(relaciones, f, indent=4, ensure_ascii=False)
```

```
import stanza
import json
from collections import Counter

# Download the Spanish language model for Stanza
stanza.download('es')

# Initialize the Stanza pipeline
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,lemma,ner,depparse')

# Texts of the biographies (Make sure doc1, doc2, and doc3 are defined)
text1 = doc1  # Democritus
text2 = doc2  # Plato
text3 = doc3  # Aristotle

texts = [text1, text2, text3]

# Process each biography
for i, text in enumerate(texts):
    print(f"Processing biography {i+1}...")

    # Annotate the text with Stanza
    doc = nlp(text)

    # Extract relationships between named entities
    relationships = []
    for sentence in doc.sentences:
        for ent1 in sentence.ents:
            for ent2 in sentence.ents:
                if ent1 != ent2:
                    # Get the index of the word within the sentence where the entity starts.
                    # We use the word's index within the sentence instead of start_char to avoid IndexError.
                    ent1_head_token_index = next(
                        (i for i, word in enumerate(sentence.words)
                         if word.start_char == ent1.start_char), None)
                    ent2_head_token_index = next(
                        (i for i, word in enumerate(sentence.words)
                         if word.start_char == ent2.start_char), None)

                    # Ensure we found valid token indices for both entities.
                    if ent1_head_token_index is not None and ent2_head_token_index is not None:
                        ent1_head_token = sentence.words[
                            ent1_head_token_index].head - 1  # Adjust index for stanza's 1-based indexing
                        ent2_head_token = sentence.words[
                            ent2_head_token_index].head - 1  # Adjust index for stanza's 1-based indexing

                        # Look for relationships based on syntactic dependencies
                        if ent1_head_token == ent2_head_token:
                            relationships.append({
                                "subject": ent1.text,
                                "relation": sentence.words[ent1_head_token].text,  # Get the head token text
                                "object": ent2.text
                            })

    # Count the frequency of relationships
    relationship_counts = Counter(tuple(rel.items()) for rel in relationships)

    # Print the extracted relationships in order of frequency (highest to lowest)
    print(f"\nRelationships between named entities for biography {i+1} (ordered by frequency):")
    for relationship, count in relationship_counts.most_common():
        relationship = dict(relationship)
        print(f" - {relationship['subject']} {relationship['relation']} {relationship['object']} (frequency: {count})")

    # Save the relationships to a JSON file
    filename = f"relationships_biography_{i+1}.json"
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(relationships, f, indent=4, ensure_ascii=False)

    print("\n")
```

```
import stanza
import json
import random
from collections import Counter

# Descargar el pipeline de stanza para inglés
stanza.download('en')

# Inicializar el pipeline en inglés
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,lemma,depparse,ner')

# Texto de ejemplo en inglés (asegúrate de que `doc3` esté definido)
text = doc1

# Procesar el texto con stanza
doc = nlp(text)

# Extraer relaciones entre entidades nombradas
relaciones = []
conteo_confianza = Counter()  # Contar la frecuencia de cada valor de confianza
for sentence in doc.sentences:
    for ent1 in sentence.ents:
        for ent2 in sentence.ents:
            if ent1 != ent2:
                # Obtener el token principal para cada entidad
                ent1_head_token_index = next((i for i, word in enumerate(sentence.words) if word.start_char == ent1.start_char), None)
                ent2_head_token_index = next((i for i, word in enumerate(sentence.words) if word.start_char == ent2.start_char), None)

                if ent1_head_token_index is not None and ent2_head_token_index is not None:
                    ent1_head_token = sentence.words[ent1_head_token_index].head - 1
                    ent2_head_token = sentence.words[ent2_head_token_index].head - 1

                    # Buscar relaciones basadas en dependencias sintácticas
                    if ent1_head_token == ent2_head_token:
                        confianza = round(random.uniform(0.5, 1), 6)  # Asignar un valor de confianza aleatorio
                        relaciones.append({
                            "confidence": confianza,
                            "subject": ent1.text,
                            "relation": sentence.words[ent1_head_token].text,
                            "object": ent2.text
                        })
                        conteo_confianza[confianza] += 1  # Actualizar el conteo de confianza

# Ordenar las relaciones por frecuencia de confianza de mayor a menor
relaciones_ordenadas = sorted(relaciones, key=lambda x: (-conteo_confianza[x["confidence"]], x["confidence"]), reverse=True)

# Imprimir las relaciones extraídas
print("Relaciones ordenadas por frecuencia de confianza:")
for relacion in relaciones_ordenadas:
    print(relacion)

# Imprimir la frecuencia de los valores de confianza
print("\nFrecuencia de los valores de confianza:")
for confianza, conteo in conteo_confianza.most_common():
    print(f"{confianza}: {conteo}")

# Guardar las relaciones en un archivo JSON
with open('relaciones_Democrito.json', 'w') as f:
    json.dump(relaciones_ordenadas, f, indent=4)
```

```
import stanza

# Descargar el modelo de Stanza para inglés
stanza.download('en')

# Inicializar el pipeline con los procesadores necesarios
nlp = stanza.Pipeline('en', processors='tokenize,mwt,pos,lemma,depparse')

def get_sentence_confidence(sentence):
    """
    Calcula un coeficiente de confianza para una oración basado en la confianza
    de sus relaciones de dependencia.

    Args:
        sentence: Una oración de Stanza.

    Returns:
        Un valor float que representa el coeficiente de confianza de la oración.
    """
    confidences = []
    for word in sentence.words:
        # Se excluyen las relaciones de puntuación y 'root'
        if word.deprel not in ['punct', 'root']:
            # Check if word.deps is not None and has at least one element
            if word.deps and len(word.deps) > 0:
                confidences.append(word.deps[0].confidence)  # Toma la confianza de la primera dependencia

    if confidences:
        return sum(confidences) / len(confidences)  # Calcula la confianza promedio
    else:
        return 0  # Si no hay dependencias, la confianza es 0


def main():
    # Ejemplo de texto
    text = doc1

    # Procesar el texto
    doc = nlp(text)

    # Obtener la confianza de cada oración
    for i, sentence in enumerate(doc.sentences):
        confidence = get_sentence_confidence(sentence)
        print(f"Sentence {i+1}: {sentence.text}")
        print(f"  Confidence: {confidence:.4f}")

if __name__ == '__main__':
    main()
```
